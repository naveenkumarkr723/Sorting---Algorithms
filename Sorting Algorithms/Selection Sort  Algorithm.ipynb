## in place comparision based algorithm
    
## 1. selection sort algorithm working (example)

### steps
    1. search the List and find out the minimum value of the List (ascending order)
    2. swap the smallest number to the 0-th index



## 2.Selection Sort Algorithm:

    1. Starting from the first element search for smallest (biggest) element in the list of numbers.

    2. Swap minimum (maximum) number with first element.

    3. take the sublist(ignore sorted part) and repeat step 1 and 2 untill all the elements are sorted


```python
list1 = [3,7,8,2,4,8,6,0]
print(list1)

length = len(list1)

for i in range(length):
    min_value = min(list1[i:])
    index_min_val = list1.index(min_value)
    
    list1[i],list1[index_min_val] = list1[index_min_val] , list1[i]
    
print(list1)   
```

```python
list2 = [3,7,8,2,4,66,66,90,8,36,0,90]
print(list2)

length = len(list2)

for i in range(length):
    min_value = max(list2[i:])
    index_min_val = list2.index(min_value,i)
    
    if list2[i] != list2.index(min_value,i):
        list2[i],list2[index_min_val] = list2[index_min_val] , list2[i]
    
print(list2)   
```

```python
### in question they are mentioned to not to use min ,max values and the index method


list2 = [3,7,8,66,2,4,66,90]
print(list2)

length = len(list2)

for i in range(length-1):
    
    ## minimum value of the list taking as its 0th index
    min_index = i
    
    ##finding the mininum value of the list
    for j in range(i+1,len(list2)):
        if list2[j] > list2[min_index]:
            min_index = j
            
    if list2[i] != list2[min_index]:
        list2[i],list2[min_index] = list2[min_index] , list2[i]
    
print("descending order",list2)   
```

```python
l = [4,5,4,3,6,2,7,1,2,1]

leng = len(l)

for i in range(leng):
    a = l[i]
    
    for j in l[i:]:    
        if a > j:
            a = j 
    
    min_val_index = l.index(a,i)
    
    #### swapping the 0 th index and min value index
    
    #if l[i] != l.index(min_val,i):
    
    l[i],l[min_val_index] = l[min_val_index],l[i]
    
```
